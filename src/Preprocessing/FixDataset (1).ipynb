```python
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/DV_A3/hundredk_spotify.csv'
df = pd.read_csv(file_path)

def process_row(row):
    last_non_null_index = row.last_valid_index()
    if last_non_null_index is None:
        return row

    last_non_null_position = df.columns.get_loc(last_non_null_index)
    if last_non_null_position >= 21:
        merge_start = 2
        merge_end = last_non_null_position - 17

        if merge_end > merge_start:
            merged_value = ','.join(row.iloc[merge_start:merge_end].dropna().astype(str))

            row.iloc[merge_start] = merged_value
            row.iloc[merge_start + 1:merge_end] = None

    print(row)
    return row

df = df.apply(process_row, axis=1)
df.to_csv('corrected_file.csv', index=False)

print("Rows processed, left-shifted, and saved to 'corrected_file.csv'")
```
